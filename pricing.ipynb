```python
queue_url_endpts = sqs.get_or_create_queue('Endpoints_Test')
vol = CloudVolume("s3://bossdb-open-data/iarpa_microns/minnie/minnie65/em", use_https=True, mip=0)
radius=(100,100,10)
resolution=(8,8,40)
ep='sqs'
save='sqs'
device='cpu'

ep_msg = sqs.get_job_from_queue(queue_url_endpts)
root_id = int(ep_msg.message_attributes['root_id']['StringValue'])
nucleus_id = int(ep_msg.message_attributes['nucleus_id']['StringValue'])
time_point = int(ep_msg.message_attributes['time']['StringValue'])
ep = np.array([float(p) for p in ep_msg.body.split(',')])
print(ep, root_id, nucleus_id, time_point, resolution)
```

```python
u=2
bound_EM  = (endpoint[0] - u*radius[0], 
            endpoint[0] + u*radius[0],
            endpoint[1] - u*radius[1],
            endpoint[1] + u*radius[1],
            endpoint[2] - radius[2],
            endpoint[2] + radius[2])
em = np.squeeze(vol[bound_EM[0]:bound_EM[1], bound_EM[2]:bound_EM[3], bound_EM[4]:bound_EM[5]])
em_vis = em[(u-1)*radius[0]:(u+1)*radius[0],
              (u-1)*radius[1]:(u+1)*radius[1], :].astype(float)
```

```python
unet_bound_mult=[1,2,3,4,5]
endpoint = np.divide(ep, resolution).astype('int')
segs = []
for u in unet_bound_mult:
    t1 = time.time()
    bound  = (endpoint[0] - radius[0], 
                endpoint[0] + radius[0],
                endpoint[1] - radius[1],
                endpoint[1] + radius[1],
                endpoint[2] - radius[2],
                endpoint[2] + radius[2])

    bound_EM  = (endpoint[0] - u*radius[0], 
                endpoint[0] + u*radius[0],
                endpoint[1] - u*radius[1],
                endpoint[1] + u*radius[1],
                endpoint[2] - radius[2],
                endpoint[2] + radius[2])
    print(bound, bound_EM)
    vol = CloudVolume("s3://bossdb-open-data/iarpa_microns/minnie/minnie65/em", use_https=True, mip=0)

    seg = np.squeeze(data_loader.get_seg(*bound))
    em = np.squeeze(vol[bound_EM[0]:bound_EM[1], bound_EM[2]:bound_EM[3], bound_EM[4]:bound_EM[5]])
    segs.append(membranes.segment_membranes(em, pth="./membrane_detection/best_metric_model_segmentation2d_dict.pth", device_s='cpu'))
    print(time.time()-t1)
```

```python
unet_bound_mult=[1,2,3,4,5]
endpoint = np.divide(ep, resolution).astype('int')
segs_thin = []
for u in unet_bound_mult:
    t1 = time.time()
    bound  = (endpoint[0] - radius[0], 
                endpoint[0] + radius[0],
                endpoint[1] - radius[1],
                endpoint[1] + radius[1],
                endpoint[2] - radius[2],
                endpoint[2] + radius[2])

    bound_EM  = (endpoint[0] - u*radius[0], 
                endpoint[0] + u*radius[0],
                endpoint[1] - u*radius[1],
                endpoint[1] + u*radius[1],
                endpoint[2] - radius[2],
                endpoint[2] + radius[2])
    print(bound, bound_EM)
    vol = CloudVolume("s3://bossdb-open-data/iarpa_microns/minnie/minnie65/em", use_https=True, mip=0)

    seg = np.squeeze(data_loader.get_seg(*bound))
    em = np.squeeze(vol[bound_EM[0]:bound_EM[1], bound_EM[2]:bound_EM[3], bound_EM[4]:bound_EM[5]])
    segs_thin.append(membranes.segment_membranes(em, pth="./membrane_detection/best_metric_model_segmentation2d_dict.pth", device_s='cpu'))
    print(time.time()-t1)
```

```python
fig, ax = plt.subplots(1,5, figsize=(15,5))

for i, u in enumerate(unet_bound_mult):
    m = segs[i]
    vis = m[(u-1)*radius[0]:(u+1)*radius[0],
                  (u-1)*radius[1]:(u+1)*radius[1], :].astype(float)
    ax[i].imshow(vis[:,:,10])
```
